```python
import pandas as pd
from textblob import TextBlob  # Import TextBlob for sentiment analysis

# Load your dataset
data = pd.read_csv('/content/tweets-data/unmag.csv')  # Replace 'your_file.csv' with your actual file path

# Check if 'full_text' column exists. If not, use another suitable column.
tweet_column = 'full_text' if 'full_text' in data.columns else data.columns[0]

# Convert the selected column into a list of tweets
data_tweet = list(data[tweet_column])

# Initialize variables
polaritas = 0
status = []
total_positif = total_negatif = total_netral = total = 0

# Sentiment analysis loop
for i, tweet in enumerate(data_tweet):
    analysis = TextBlob(tweet)
    polaritas += analysis.polarity

    if analysis.sentiment.polarity > 0.0:
        total_positif += 1
        status.append('Positif')
    elif analysis.sentiment.polarity == 0.0:
        total_netral += 1
        status.append('Netral')
    else:
        total_negatif += 1
        status.append('Negatif')

    total += 1

# Print results
print(f'Hasil Analisis Data:\nPositif = {total_positif}\nNetral = {total_netral}\nNegatif = {total_negatif}')
print(f'\nTotal Data : {total}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob

# Check if 'full_text' column exists. If not, use another suitable column.
tweet_column = 'full_text' if 'full_text' in data.columns else data.columns[0]

data_tweet = list(data[tweet_column])
polaritas = 0

status = []
total_positif = total_negatif = total_netral = total = 0

for i, tweet in enumerate(data_tweet):
    analysis = TextBlob(tweet)
    polaritas += analysis.polarity

    if analysis.sentiment.polarity > 0.0:
        total_positif += 1
        status.append('Positif')
    elif analysis.sentiment.polarity == 0.0:
        total_netral += 1
        status.append('Netral')
    else:
        total_negatif += 1
        status.append('Negatif')

    total += 1

# Create a DataFrame to show the results
sentiment_data = {
    'Sentiment': ['Positif', 'Netral', 'Negatif'],
    'Count': [total_positif, total_netral, total_negatif]
}

df_sentiment = pd.DataFrame(sentiment_data)

# Display the table
print(f'Hasil Analisis Data:\n{df_sentiment}')
print(f'\nTotal Data : {total}')

# Plotting the results
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(df_sentiment['Sentiment'], df_sentiment['Count'], color=['green', 'gray', 'red'])
ax.set_xlabel('Sentiment')
ax.set_ylabel('Count')
ax.set_title('Sentiment Analysis Results')

plt.show()
```

```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob

# File path
file_path = 'tweets-data/unmag.csv'

# Title of the app
st.title("Analisis Sentimen Tweet - Patrick Kluivert")

# Load data
df = pd.read_csv(file_path)

# Display raw data
st.header("Data Tweet Mentah")
st.write(df)

# Sentiment Analysis
st.header("Analisis Sentimen")

# Normalisasi fungsi
norm = {
    " yg ": " yang ", " towel ": " ", " erick tohir ": " ",
    ' nggak ': ' tidak ', ' gak ': ' tidak ', ' pssi ': ' banget dari ',
    # tambahkan normalisasi lainnya jika diperlukan
}

def normalisasi(str_text):
    for i in norm:
        str_text = str_text.replace(i, norm[i])
    return str_text

df['normalized_text'] = df['full_text'].apply(normalisasi)

# Sentiment analysis
polaritas = 0
status = []
total_positif = total_negatif = total_netral = 0

for tweet in df['normalized_text']:
    analysis = TextBlob(tweet)
    polaritas += analysis.polarity

    if analysis.sentiment.polarity > 0.0:
        total_positif += 1
        status.append('Positif')
    elif analysis.sentiment.polarity == 0.0:
        total_netral += 1
        status.append('Netral')
    else:
        total_negatif += 1
        status.append('Negatif')

df['Sentiment'] = status

# Display sentiment analysis results
st.subheader("Hasil Analisis Sentimen")
st.write(f"Positif: {total_positif}, Netral: {total_netral}, Negatif: {total_negatif}")
st.write(df[['normalized_text', 'Sentiment']])

# Plot sentiment analysis
sentiment_data = {
    'Sentiment': ['Positif', 'Netral', 'Negatif'],
    'Count': [total_positif, total_netral, total_negatif]
}
df_sentiment = pd.DataFrame(sentiment_data)

fig, ax = plt.subplots()
ax.bar(df_sentiment['Sentiment'], df_sentiment['Count'], color=['green', 'gray', 'red'])
ax.set_title("Distribusi Sentimen")
ax.set_xlabel("Sentimen")
ax.set_ylabel("Jumlah")
st.pyplot(fig)

st.write("Aplikasi ini membantu menampilkan hasil analisis tweet terkait Patrick Kluivert secara langsung di web.")
```
